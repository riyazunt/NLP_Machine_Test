```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, TfidfTransformer
```

```python
cv = CountVectorizer(stop_words='english')
```

```python
cvm = cv.fit_transform(df['review_by_patient'])
```

```python
cvm
```

```python
cvm.todense()
```

```python
cv.vocabulary_
```

```python
tfidf = TfidfTransformer()
```

```python
results = tfidf.fit_transform(cvm) # bag of words to tfidf freequency count
```

```python
results
```

```python
resmat = results.todense()
```

```python
resmat
```

```python
resmat[0].max()
```
